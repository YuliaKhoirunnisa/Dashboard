## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Pengaruh Musim dan Cuaca Terhadap Jumlah Penyewaan Sepeda?

```python
# Menghitung total penyewaan sepeda berdasarkan musim dan cuaca
bike_rentals_season_weather = hour_df.groupby(["season", "weathersit"])["cnt"].sum().reset_index()

# Mengubah kode musim menjadi label
season_labels = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
weather_labels = {1: "Clear", 2: "Cloudy", 3: "Rain/Snow"}

bike_rentals_season_weather["season"] = bike_rentals_season_weather["season"].map(season_labels)
bike_rentals_season_weather["weathersit"] = bike_rentals_season_weather["weathersit"].map(weather_labels)

# Membuat plot dengan Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=bike_rentals_season_weather, x="season", y="cnt", hue="weathersit", palette="coolwarm")

# Menambahkan judul dan label
plt.title("Pengaruh Musim dan Cuaca terhadap Jumlah Penyewaan Sepeda", fontsize=14)
plt.xlabel("Musim", fontsize=12)
plt.ylabel("Total Penyewaan", fontsize=12)
plt.legend(title="Kondisi Cuaca")

plt.show()
```

### Pertanyaan 2: Pada Jam Berapa Peminat Penyewaan Sepeda TertinggiÂ danÂ Terendah?

```python
df = pd.read_csv("hour.csv")

# Mengelompokkan data berdasarkan jam dan menghitung total penyewaan sepeda
bike_rentals_per_hour = df.groupby("hr")["cnt"].sum()

# Mencari jam dengan penyewaan tertinggi dan terendah
max_hour = bike_rentals_per_hour.idxmax()
max_rentals = bike_rentals_per_hour.max()
min_hour = bike_rentals_per_hour.idxmin()
min_rentals = bike_rentals_per_hour.min()

# Membuat figure untuk visualisasi
plt.figure(figsize=(12, 6))

# Line plot jumlah penyewaan per jam
sns.lineplot(x=bike_rentals_per_hour.index, y=bike_rentals_per_hour.values, marker="o", label="Penyewaan Sepeda")

# Menandai jam dengan penyewaan tertinggi dan terendah
plt.scatter(max_hour, max_rentals, color='red', s=100, label=f"Max: {max_hour}:00 ({max_rentals})")
plt.scatter(min_hour, min_rentals, color='blue', s=100, label=f"Min: {min_hour}:00 ({min_rentals})")

# Menambahkan keterangan
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.title("Jumlah Penyewaan Sepeda per Jam")
plt.xticks(range(0, 24))  # Menampilkan semua jam dari 0 sampai 23
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```

**Rekomendasi:**  
1. **Peningkatan layanan selama musim gugur** â€“ Operator penyewaan sepeda dapat menambahkan lebih banyak sepeda dan meningkatkan fasilitas di musim gugur karena permintaan tinggi.  
2. **Strategi promosi di musim semi dan musim dingin** â€“ Menawarkan diskon atau promo khusus untuk menarik lebih banyak penyewa di musim dengan permintaan rendah.  
3. **Penyediaan fasilitas pelindung cuaca** â€“ Menyediakan stasiun parkir yang lebih terlindungi dan sepeda dengan aksesori tahan cuaca agar lebih nyaman digunakan dalam kondisi mendung atau hujan.  

**Pertanyaan 2: Jam Peminat Penyewaan Sepeda Tertinggi dan Terendah**  
- **Jumlah penyewaan tertinggi** terjadi pada pukul **17:00 dengan 336.860 penyewaan**. Ini menunjukkan bahwa banyak orang menggunakan sepeda sebagai sarana transportasi setelah jam kerja atau untuk beraktivitas sore hari.

**Rekomendasi:**  
1. **Penyesuaian jumlah sepeda berdasarkan jam sibuk** â€“ Menyediakan lebih banyak sepeda menjelang sore hari (16:00-18:00) untuk mengakomodasi lonjakan permintaan.
